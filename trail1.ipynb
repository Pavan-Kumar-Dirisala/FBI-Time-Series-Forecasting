```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_excel('Train.xlsx')
train_df.head()
```

```python
test_df = pd.read_csv('Test (2).csv')
test_df.head()
```

```python
train_df.isnull().sum()
import pandas as pd

# Ensure Date column is in datetime format
train_df['Date'] = pd.to_datetime(train_df[['YEAR', 'MONTH', 'DAY']])

# Step 1: Find the last day of each month present in the dataset
last_days = train_df.groupby(train_df['Date'].dt.to_period('M'))['Date'].max().reset_index(drop=True)

# Step 2: Filter original data to keep only rows where Date is in last_days
filtered_df = train_df[train_df['Date'].isin(last_days)]

# Step 3: Group by Date and TYPE, count incidents
incident_summary = filtered_df.groupby(['Date', 'TYPE']).size().reset_index(name='incident_donut')

# View result
incident_summary
# Ensure Date is datetime in train_df
train_df['Date'] = pd.to_datetime(train_df[['YEAR', 'MONTH', 'DAY']])

# Aggregate monthly incident counts by YEAR, MONTH, TYPE
monthly_counts = train_df.groupby(['YEAR', 'MONTH', 'TYPE']).size().reset_index(name='Incident_Counts')

# To view the data similar to your test data
print("Test_monthly")
print(monthly_counts.head())
print("Train_Monthly")
print(test_df.head())
print("Incident_summary")
print(incident_summary.head())
```
